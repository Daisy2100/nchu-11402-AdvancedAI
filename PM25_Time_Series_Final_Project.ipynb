````python
import os
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

PROJECT_ROOT = r"C:\Users\wecan\OneDrive\獢\?刻撠\鞈極\1142_銝玨\銝\??勗?"

data_dir = os.path.join(PROJECT_ROOT, "data")
candidate_files = ["air_quality.csv", "taiwan_air_quality.csv"]
DATA_PATH = None
for file_name in candidate_files:
    candidate_path = os.path.join(data_dir, file_name)
    if os.path.exists(candidate_path):
        DATA_PATH = candidate_path
        break

if DATA_PATH is None:
    available_files = os.listdir(data_dir) if os.path.isdir(data_dir) else []
    raise FileNotFoundError(
        f"Cannot find dataset in {data_dir}. Tried: {candidate_files}. Available: {available_files}"
    )
OUTPUT_ROOT = os.path.join(PROJECT_ROOT, "output")
FIGURE_DIR = os.path.join(OUTPUT_ROOT, "figures")
RESULT_DIR = os.path.join(OUTPUT_ROOT, "results")

os.makedirs(FIGURE_DIR, exist_ok=True)
os.makedirs(RESULT_DIR, exist_ok=True)

print("DATA_PATH:", DATA_PATH)
print("OUTPUT_ROOT:", OUTPUT_ROOT)
````

## 2. Load Dataset

霈????嚗?瑼Ｘ鞈?憭批???雿?蝔梯??嗾蝑??n

````python
df = pd.read_csv(DATA_PATH)

print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
display(df.head())
````

## 3. Auto-detect Time and PM2.5 Columns

?挾??皜砍虜閬???甈???PM2.5 甈??n
憒?憭望?嚗???銝?潸撓?箇?甈??迂嚗???摰?`time_col` ??`pm25_col`?n

````python
possible_time_cols = [
    "DataCreationDate", "PublishTime", "Date", "date",
    "datetime", "Datetime", "time", "Time"
]

possible_pm25_cols = [
    "PM2.5", "PM25", "pm25", "pm2.5", "PM2_5", "pm2_5"
]

time_col = None
pm25_col = None

for col in df.columns:
    if col in possible_time_cols:
        time_col = col
        break

for col in df.columns:
    if col in possible_pm25_cols:
        pm25_col = col
        break

print("Detected datetime column:", time_col)
print("Detected PM2.5 column:", pm25_col)

if time_col is None:
    raise ValueError("Cannot find datetime column. Please manually assign time_col.")

if pm25_col is None:
    raise ValueError("Cannot find PM2.5 column. Please manually assign pm25_col.")
````

## 4. Data Cleaning

皜??批捆?嚗n
- 頧???甈?
- 頧? PM2.5 ?箸?墦n- 蝘駁蝻箏仃?墦n- ?亥???憭葫蝡???洵銝?葫蝡脰??桃???摨???
- 蝘駁銝???鞎墦n- 撠???? hourly time series
- 雿輻???扳?鋆撩憭勗墦n

````python
df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
df[pm25_col] = pd.to_numeric(df[pm25_col], errors="coerce")

# Remove rows without datetime or PM2.5
df = df.dropna(subset=[time_col, pm25_col])
df = df.sort_values(time_col)

# If station column exists, select one station to avoid mixing different locations
site_col = None
possible_site_cols = ["SiteName", "site", "Site", "sitename", "皜祉?"]

for col in df.columns:
    if col in possible_site_cols:
        site_col = col
        break

if site_col is not None:
    print("Available stations:")
    print(df[site_col].dropna().unique()[:20])
    selected_site = df[site_col].dropna().unique()[0]
    print("\nSelected station:", selected_site)
    df = df[df[site_col] == selected_site].copy()

# Keep only required columns
df = df[[time_col, pm25_col]].copy()
df.columns = ["datetime", "pm25"]

# Remove abnormal values
df = df[df["pm25"] >= 0]

# Set datetime index and handle duplicate timestamps
df = df.set_index("datetime").sort_index()
df = df.groupby(df.index).mean()

# Resample to hourly time series
df = df.resample("H").mean()

# Fill missing values
df["pm25"] = df["pm25"].interpolate(method="time")
df["pm25"] = df["pm25"].fillna(method="bfill").fillna(method="ffill")

print("Cleaned dataset shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())

display(df.head())
````

## 6. Feature Extraction

?箔???璅∪?銵函嚗遣蝡誑銝????敺蛛?

- Time features: hour, day of week, month, weekend
- Cyclical encoding: hour sine/cosine
- Lag features: previous 1, 2, 3, 6, 12, 24 hours
- Rolling mean features: previous 3, 6, 24 hours
- Target: next-hour PM2.5


````python
# ============================================================
# 6. FEATURE ENGINEERING
# ============================================================

print("\nPerforming feature engineering...")

data = df.copy()

# ============================================================
# Time Features
# ============================================================

data["hour"] = data.index.hour
data["dayofweek"] = data.index.dayofweek
data["month"] = data.index.month

data["is_weekend"] = (
    data["dayofweek"] >= 5
).astype(int)

# ============================================================
# Cyclical Encoding
# ============================================================

data["hour_sin"] = np.sin(
    2 * np.pi * data["hour"] / 24
)

data["hour_cos"] = np.cos(
    2 * np.pi * data["hour"] / 24
)

data["month_sin"] = np.sin(
    2 * np.pi * data["month"] / 12
)

data["month_cos"] = np.cos(
    2 * np.pi * data["month"] / 12
)

# ============================================================
# Lag Features
# ============================================================

for lag in [1, 2, 3, 6, 12, 24]:

    data[f"pm25_lag_{lag}"] = (
        data["pm25"].shift(lag)
    )

# ============================================================
# Rolling Mean Features
# ============================================================

data["rolling_3"] = (
    data["pm25"]
    .shift(1)
    .rolling(3)
    .mean()
)

data["rolling_6"] = (
    data["pm25"]
    .shift(1)
    .rolling(6)
    .mean()
)

data["rolling_24"] = (
    data["pm25"]
    .shift(1)
    .rolling(24)
    .mean()
)

# ============================================================
# Prediction Target
# ============================================================

# ?葫銝?撠? PM2.5

data["target"] = data["pm25"].shift(-1)

# remove NA
data = data.dropna()

print("\nFeature engineered dataset shape:")
print(data.shape)

# ============================================================
# Feature Columns
# ============================================================

feature_cols = [

    "hour",
    "dayofweek",
    "month",
    "is_weekend",

    "hour_sin",
    "hour_cos",

    "month_sin",
    "month_cos",

    "pm25_lag_1",
    "pm25_lag_2",
    "pm25_lag_3",

    "pm25_lag_6",
    "pm25_lag_12",
    "pm25_lag_24",

    "rolling_3",
    "rolling_6",
    "rolling_24"
]

# ============================================================
# 6.5 PCA (Dimension Reduction)
# ============================================================

from sklearn.decomposition import PCA

print("\nPerforming PCA...")

# prepare data
feature_for_pca = data[feature_cols].copy()

# standardize
scaler_pca = StandardScaler()

X_scaled = scaler_pca.fit_transform(
    feature_for_pca
)

# PCA
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

# explained variance
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

# PCA Visualization
plt.figure(figsize=(8, 6))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    alpha=0.5
)

plt.title("PCA Visualization")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.tight_layout()

plt.savefig(
    os.path.join(
        FIGURE_DIR,
        "pca_visualization.png"
    ),
    dpi=300
)

plt.close()

# save PCA result
pca_df = pd.DataFrame({

    "PC1": X_pca[:, 0],

    "PC2": X_pca[:, 1]
})

pca_df.to_csv(
    os.path.join(
        RESULT_DIR,
        "pca_result.csv"
    ),
    index=False
)

# ============================================================
# 6.6 KMeans Clustering
# ============================================================

from sklearn.cluster import KMeans

print("\nPerforming KMeans clustering...")

kmeans = KMeans(

    n_clusters=3,

    random_state=42,

    n_init=10
)

clusters = kmeans.fit_predict(X_scaled)

# add cluster label
data["cluster"] = clusters

print("\nCluster counts:")
print(
    data["cluster"].value_counts()
)

# cluster visualization
plt.figure(figsize=(8, 6))

scatter = plt.scatter(

    X_pca[:, 0],

    X_pca[:, 1],

    c=clusters,

    alpha=0.6
)

plt.title(
    "KMeans Clustering on PCA Features"
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.colorbar(scatter)

plt.tight_layout()

plt.savefig(
    os.path.join(
        FIGURE_DIR,
        "kmeans_clustering.png"
    ),
    dpi=300
)

plt.close()

# save clustering result
cluster_df = pd.DataFrame({

    "PC1": X_pca[:, 0],

    "PC2": X_pca[:, 1],

    "Cluster": clusters
})

cluster_df.to_csv(
    os.path.join(
        RESULT_DIR,
        "kmeans_cluster_result.csv"
    ),
    index=False
)
````

## 7. Modeling and Verification Split

靘雿平閬?嚗蝙?冽????? 80% 雿 modeling/training data嚗?敺?20% 雿 verification/testing data?n

````python
feature_cols = [
    "hour", "dayofweek", "month", "is_weekend",
    "hour_sin", "hour_cos",
    "pm25_lag_1", "pm25_lag_2", "pm25_lag_3",
    "pm25_lag_6", "pm25_lag_12", "pm25_lag_24",
    "rolling_3", "rolling_6", "rolling_24"
]

X = data[feature_cols]
y = data["target"]

split_idx = int(len(data) * 0.8)

X_train = X.iloc[:split_idx]
X_test = X.iloc[split_idx:]
y_train = y.iloc[:split_idx]
y_test = y.iloc[split_idx:]

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
print("Train period:", X_train.index.min(), "to", X_train.index.max())
print("Test period:", X_test.index.min(), "to", X_test.index.max())
````

## 8. Modeling

撱箇??拙芋?脰?瘥?嚗n
1. Linear Regression: baseline model
2. Random Forest Regressor: nonlinear machine learning model


````python
models = {
    "Linear Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    "Random Forest": RandomForestRegressor(
        n_estimators=200,
        max_depth=12,
        random_state=42,
        n_jobs=-1
    )
}

results = {}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    results[name] = {
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2,
        "Prediction": pred
    }

    print("MAE :", round(mae, 4))
    print("RMSE:", round(rmse, 4))
    print("R2  :", round(r2, 4))
    print("-" * 40)
````

## 9. Model Evaluation Results

雿輻 MAE?MSE?簡 閰摯璅∪?銵函?n

````python
metrics_df = pd.DataFrame(results).T.drop(columns=["Prediction"])
display(metrics_df)

metrics_df.to_csv(os.path.join(RESULT_DIR, "model_metrics.csv"), index=True)
````

## 12. Discussion Template

### Problem Definition
This project aims to predict the next-hour PM2.5 concentration using historical air quality time series data. The task is formulated as a time series regression problem.

### Data Cleaning
The raw dataset was cleaned by converting the timestamp column into datetime format, converting PM2.5 values into numeric format, removing missing or abnormal values, selecting a single monitoring station, resampling the data into hourly frequency, and filling missing values using time-based interpolation.

### Feature Extraction
To improve prediction performance, several time series features were extracted, including time-based features, cyclical hour encoding, lag features, and rolling mean features. Lag and rolling features help the model capture short-term and daily temporal dependency.

### Modeling
Two models were compared: Linear Regression and Random Forest. Linear Regression was used as a baseline model, while Random Forest was used to capture nonlinear patterns in the time series.

### Verification
Following the project requirement, the first 80% of the time series was used for modeling, and the last 20% was used for verification. Model performance was evaluated using MAE, RMSE, and R簡.

### Result Discussion
If Random Forest outperforms Linear Regression, it suggests that PM2.5 concentration has nonlinear temporal patterns. Feature importance can be used to explain whether recent PM2.5 values or daily patterns are more influential in prediction.
